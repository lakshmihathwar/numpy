```python
import numpy as np
```

```python
def basic_math(array1,array2):
  print(array1)
  print(array2)

  print("\n")

  add = array1 + array2
  sub = array1 - array2
  mul = array1 * array2
  div = array1 / array2
  pow = array1 ** array2

  print("Addition of two arrays:",add)
  print("Subtraction of two arrays:",sub)
  print("Multiplication of two arrays:",mul)
  print("Division of two arrays:",div)
  print("Power Operation of two arrays:",pow)
```

```python
def avg(arr):
  return np.mean(arr)

def median(arr):
  return np.median(arr)

def std(arr):
  return np.std(arr)

def var(arr):
  return np.var(arr)

def maxof(arr):
  return np.max(arr)

def minof(arr):
  return np.min(arr)

def sumof(arr):
  return np.sum(arr)

def prodof(arr):
  return np.prod(arr)
```

```python
def stats_operations(array1,array2):
  print(array1)
  print(array1)

  print()

  print("Mean of first array:",avg(array1))
  print("Mean of second array:",avg(array2))

  print("\n")

  print("Median of first array:",median(array1))
  print("Median of second array:",median(array2))

  print("\n")

  print("Standard Deviation of first array:",std(array1))
  print("Standard Deviation of second array:",std(array2))

  print("\n")

  print("Variance of first array:",var(array1))
  print("Variance of second array:",var(array2))

  print("\n")

  print("Sum of first array:",sum(array1))
  print("Sum of second array:",sum(array2))

  print("\n")

  print("Product of first array:",prodof(array1))
  print("Product of second array:",prodof(array2))

  print("\n")

  print("Maximum of first array:",max(array1))
  print("Maximum of second array:",max(array2))

  print("\n")

  print("Minimum of first array:",min(array1))
  print("Minimum of second array:",min(array2))

  print("\n")

  print("Shape of the array is:",array1.shape)
  print("Shape of the array is:",array2.shape)

  print("\n")

  print("Size of the array is:",array1.size)
  print("Size of the array is:",array2.size)

  print("\n")

  print("Datatype of the array is:",array1.dtype)
  print("Datatype of the array is:",array2.dtype)

  print("\n")
```

```python
def arrReshape(array):
  print("1D to 2D Array")
  for i in range(1, array.size + 1):
        if array.size % i == 0:
            row, col = i, array.size // i
            print("\nShape:",(row,col))
            print("Array \n", array.reshape(row, col))

  print("\n")

  print("1D to 3D Array")
  for i in range(1, array.size + 1):
      for j in range(1, array.size + 1):
          if array.size % (i * j) == 0:
              k = array.size // (i * j)
              print("\nShape:",(i,j,k))
              print("Array \n", array.reshape(i, j, k))


def sort(arr):
  return np.sort(arr)

def rev(arr):
  return np.sort(arr)[::-1]

def unique(arr):
  return np.unique(arr)

def extract(arr):
  # arr = np.sort(arr)
  li = [float(arr[0]),float(arr[-1])]
  return li

def filter(arr):
  return arr[arr > np.mean(arr)]
```

```python
def array_manipulation(array1,array2):
    print(array1)
    print(array1)

    print()

    print("Possible reshapes for the provided array:\n")
    print("\nArray 1:")
    arrReshape(array1)
    print("\nArray 2")
    arrReshape(array2)

    print("\n")

    print("Sorted first array:",sort(array1))
    print("Sorted second array:",sort(array2))

    print("\n")

    print("Reverse Sorting first array:",rev(array1))
    print("Reverse Sorting second array:",rev(array2))

    print("\n")

    print("Unique elements of first array:",unique(array1))
    print("Unique elements of second array:",unique(array2))

    print("\n")

    print("First and last element of first array:",extract(array1))
    print("First and last element of second array:",extract(array2))

    print("\n")

    print("Elements greater than mean of first array:",filter(array1))
    print("Elements greater than mean of second array:",filter(array2))

```

```python
def get_arrays():
    while True:
        print("Please, enter only numbers!")
        arr1 = input("Enter the first array (each value separated by spaces): ")
        arr2 = input("Enter the second array (each value separated by spaces): ")
        array1 = np.array(arr1.split(), dtype=float)
        array2 = np.array(arr2.split(), dtype=float)
        print()
        if array1.size != array2.size:
            print("The arrays must be of the same size. Please enter again.\n")
        else:
            return array1, array2

def is_valid_choice(choice):
    if choice not in ['1', '2', '3', '4']:
        return False
    return True

def enter():
    print()
    input("Press Enter to continue...")
    print()

while True:
    print('=' * 50)
    print("          Numpy Mathematical Calculator            ")
    print('=' * 50)
    print("1. Basic Arithmetic Operations")
    print("2. Statistical Operatios")
    print("3. Array Manipulation")
    print("4. Exit")
    print("=" * 50)
    choice = input("Your choice: ")

    if not is_valid_choice(choice):
        print("Invalid input. Please enter 1, 2, 3, or 4.")
        continue

    n = int(choice)

    if n == 4:
        print("Thank you!")
        break

    array1, array2 = get_arrays()

    if n == 1:
        basic_math(array1, array2)
        enter()
    elif n == 2:
        stats_operations(array1, array2)
        enter()
    elif n == 3:
        array_manipulation(array1, array2)
        enter()
```
